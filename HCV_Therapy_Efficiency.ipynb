```python
def Fisolate_source(record):
    isolation_source=['isolation_source']
    country=['country']
    version=['version']
   
    for l in record:
        if l.startswith('VERSION'):
            version.append(l[12:])
            
        elif l.startswith('/country='):
            country.append(l[10:-1])
            
        elif l.startswith('/isolation_source'):
                    isolation_source.append(l[19:-1])
                
    return (version,isolation_source,country ) 
```
